```python
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import cv2
import time
import torch
# Load the trained YOLOv5 model
model = YOLO('/content/drive/My Drive/Models/Yolov5l_4labels.pt')  # Replace with your trained model path
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
model = model.to(device)
# Run inference on a test image
test_image_path = '/content/drive/MyDrive/Yolov5_4labels/dataset/images/val/frame_0562.jpg' # Replace with your test image path
#test_image_path = 'image8.jpg'
print('here')
image = cv2.imread(test_image_path)

start_time = time.time()
results = model.predict(source=test_image_path, conf=0.5)  # Run inference with a confidence threshold of 0.5
end_time = time.time()
time_taken = end_time - start_time
print("Time required to show the output:", time_taken, "seconds")
coordinates =[]
# Extract bounding box coordinates, class labels, and confidence scores
for result in results:
    boxes = result.boxes.xyxy  # Bounding box coordinates in [x1, y1, x2, y2] format
    class_ids = result.boxes.cls  # Class IDs
    scores = result.boxes.conf  # Confidence scores

    # Print each detection
    for i in range(len(boxes)):
        x1, y1, x2, y2 = boxes[i]  # Coordinates of the bounding box
        class_id = int(class_ids[i])  # Class ID
        score = scores[i].item()  # Confidence score

        coordinates.append((int((x1+x2)/2),int((y1+y2)/2)))
        for (x, y) in coordinates:
            # Draw a small circle at each coordinate
            cv2.circle(image, (x, y), radius=5, color=(0, 255, 0), thickness=-1)
            # Optionally, add text near each coordinate
            cv2.putText(image, f'({x}, {y})', (x + 10, y - 10),
                        fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.5,
                        color=(0, 255, 0), thickness=1)
cv2_imshow(image)
# Display the image

        # print(f"Detection {i+1}:")
        # print(f"  Class ID: {class_id}")
        # print(f"  Confidence Score: {score:.2f}")
        # print(f"  Bounding Box: [{x1:.2f}, {y1:.2f}, {x2:.2f}, {y2:.2f}]\n")
```

```python
from ultralytics import YOLO
import torch
import numpy as np

# Load the trained YOLOv5 model
#model = YOLO('runs/train/stick_detection/weights/best.pt')  # Replace with your trained model path

# Load validation data
data_yaml = '/content/drive/My Drive/Yolov5_4labels/data.yaml'  # Your data.yaml file

# Evaluate the model on the validation set
metrics = model.val(data=data_yaml, conf=0.5, iou=0.5)  # Set confidence and IoU thresholds as desired

# Access mean precision and mean recall as attributes
mean_precision = np.mean(metrics.box.p)  # Mean precision across all classes
mean_recall = np.mean(metrics.box.r)     # Mean recall across all classes

# Calculate a custom "accuracy-like" metric as the harmonic mean of precision and recall
if mean_precision + mean_recall > 0:
    approximate_accuracy = 2 * (mean_precision * mean_recall) / (mean_precision + mean_recall)
else:
    approximate_accuracy = 0

# Print results
print(f"Mean Precision: {mean_precision:.4f}")
print(f"Mean Recall: {mean_recall:.4f}")
print(f"Approximate Accuracy: {approximate_accuracy:.4f}")
```
